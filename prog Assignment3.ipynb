```python
# 1.
i = int(input())
if i>0:
    print("Positive")
elif i <0:
    print("Negative")
else:
    print("Zero")
```

```python
# 2.
i = int(input())
if i % 2==0:
    print("Even")
else:
    print("Odd")
    
```

```python
# 3.
year = int(input())
if year %4 ==0 and year %400==0:
    print("leap year")
else:
    print("Not leap year")
```

```python
# 4.
def checkYear(year):
    return (((year % 4 == 0) and (year % 100 != 0)) or (year % 400 == 0));
 
year = 2041
if(checkYear(year)):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
# 5.
for i in range(1,10000):
    if i % 2 !=0:
        print(i,"prime number")
else:
    print(i,"Not prime number")
        
```
